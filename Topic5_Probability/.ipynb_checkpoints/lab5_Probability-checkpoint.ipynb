## Itertools
The code below shows all the permutations of 3 strings A, B, C

```python
import itertools
strList = ['A', 'B', 'C']
result = list(itertools.permutations(strList))
for item in result:
    print(item)
```

```python
#What will this code print?
strList = ['A','B','C','D','E']
result = list(itertools.permutations(strList))
#resultLength = len(list(result)) #list() converts the itertools.permutations object to a list. len() gets the length of the list
#print(f"There are a total of {resultLength} permutations")
for item in result:
    print(item)
```

** In how many ways can 3 lifeguards be chosen and ordered among the stations?**

```python
strList = ['A','B','C','D','E']
result = itertools.permutations(strList,3) #The second argument of the function specifies number of objects to select
resultLength = len(list(result)) 
print(f"The total number of permutations of 3 lifeguards chosen from 5 is {resultLength}")
list(itertools.permutations(strList,3))
```

**The number of ways of choosing any 3 lifeguards from a group of 5 life guards**

```python
strList = ['A','B','C','D','E']
resultList = list(itertools.combinations(strList,3)) #The second argument of the function specifies number of objects to select
print(f"The total number of ways of choosing 3 lifeguards from 5 is {len(resultList)}")

#You can store all of these outcomes in a pandas DataFrame as shown below:
import pandas as pd
df = pd.DataFrame({'outcomes':resultList})
df
```

## Probability Calculations
* The first example below shows how to count the total number of outcomes of rolling a die once

```python
numList = list(range(1,7)) #gives a list from 1 to 6 representing the total sample space
resultList = list(itertools.combinations(numList,1)) #This does 6 Combination 1, to basically choose 1 out of 6 outcomes
#resultList = list(itertools.permutations(numList,1)) #Gives same result as combinations when choosing just 1.
print(f"Total number of outcomes = {len(resultList)}")
print(resultList)
```

**The code below illustrates how to compute the probabililities of all possible total values that can be obtained from rolling a fair die two times (or rolling two fair dice)**

```python
numList = list(range(1,7))
resultList1 = list(itertools.product(numList,repeat=2)) #this is how we simulate rolling the die twice (or rolling 2 dice)
print(f"Total number of outcomes = {len(resultList1)}")
#print(resultList1)

import numpy as np
resultArray = np.array(resultList1)
sumOutcomes = np.sum(resultArray, axis=1)

df1 = pd.DataFrame({"outcomes":resultList1,"sumOutcomes":sumOutcomes})
df1
```

```python
by_sumOutcomes = df1.groupby("sumOutcomes")
by_sumOutcomes.describe()
outcomeDF = by_sumOutcomes.count()
outcomes = outcomeDF.values[:,0]
probability = outcomes / np.sum(outcomes)
newDF = pd.DataFrame({"sum":df1["sumOutcomes"].unique(), "outcomes":outcomes, "probability":probability})
newDF
```
